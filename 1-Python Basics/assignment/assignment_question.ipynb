**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
user_input = int(input("Enter number to calculate it's factorial: "))
print(range(1,user_input))
factorial = 1
for i in range(1,user_input+1):
    factorial *= i
print(factorial)
```
