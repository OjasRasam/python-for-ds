##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
## why functions?
num=-1
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def odd_or_even(num):
  """This function takes in a number and prints if it it odd or even"""
  if num % 2 == 0:
    print("Even")
  else:
    print("Odd")
```

```python
## Call this function
odd_or_even(22)
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

print(add(5,3))
    
```

```python
## Default Parameters

def greet(name="Guest"):
    print(f"Hello {name} Welcome to the paradise")

greet("Ojas")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome To the paradise")

greet("Krish")
```

```python
### Variable Length Arguments
### Positional And Kwargs(Keyword Arguments)


## Positional Arguments
def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Ojas")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Ojas")
```

```python
### Keywords Arguments

def print_details(**kwargs):
    ### Parms are in key : value pairs
    for key, value in kwargs.items():
        print(f"{key}: {value}")
    
```

```python
print_details(name="Ojas", age=24, loction="Thakurli")
```

```python
def print_details(*args, **kwargs):
    for num in args:
        print(num)
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(1,2,3,4,5,name="Ojas", age=24)
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
